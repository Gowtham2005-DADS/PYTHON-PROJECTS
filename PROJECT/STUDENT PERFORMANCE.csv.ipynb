# **HYPOTHESIS TESTING**

^^ **ONE-SAMPLE T-TEST**

-SCIPY.STATS--

```python
import scipy.stats as stats
```

```python
Math_Scores = student['Math Score']

t_stat, p_value = stats.ttest_1samp(Math_Scores, 70)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H₀ = Mean math score is significantly different from 70")
else:
    print("Fail to reject H₀ = No significant difference from 70")
```

^**INSIGHTS**

  * THE TEST CHECKS IF THE MEAN MATH SCORE IS 70

  * **P-VALUES** IS **MUCH LESS THAN 0.05** - **REJECT HYPOTHESIS MEAN**

  * MEAN **MATH SCORE** IS **SIGNIFICANTLY DIFFERENT** FROM **70**

^^ **INDEPENDENT TWO-SAMPLE T-TEST**

```python
male_scores = student[student['Gender'] == 'male']['Math Score']
female_scores = student[student['Gender'] == 'female']['Math Score']

t_stat, p_value = stats.ttest_ind(male_scores, female_scores)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H₀ = Significant difference between male & female math scores")
else:
    print("Fail to reject H₀ = No significant difference")
```

^ **INSIGHTS**
   * THIS CODE IS A **TWO-SAMPLE T-TEST** COMPARING MALE AND FEMALE MATH SCORE

   * **P-VALUE** IS FAR **BELOW 0.05** - REJECT HYPOTHESIS MEAN

   * THERE IS A **SIGNIFICANT DIFFERENCE** BETWEEN **MALE AND FEMALE MATH SCORES**

^^ **PAIRED T-TEST**

```python
reading_scores = student['Reading Score']
writing_scores = student['Writing Score']

t_stat, p_value = stats.ttest_rel(reading_scores, writing_scores)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H₀ = Significant difference between reading & writing scores")
else:
    print("Fail to reject H₀ = No significant difference")
```

^ **INSIGHTS**
   * THIS CODE IS **PAIRED T-TEST** COMPARING **READING** AND **WRITING SCORES** OF THE **SAME STUDENT**

   * **P-VALUES** IS **EXTREMELY SMALL(<0.05)**-**REJECT NULL HYPOTHESIS**

   * THERE IS A **SIGNIFICANT DIFFERENCE** BETWEEN **READING** AND **WRITING SCORES**

^^ **ONE-SAMPLE Z-TEST**

```python
from statsmodels.stats.weightstats import ztest
```

1. **Check if mean Reading score = 70**

```python
Reading_Scores = student['Reading Score']
z_stat, p_value = ztest(Reading_Scores, value=70)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H₀ = Mean math score is significantly different from 70")
else:
    print("Fail to reject H₀ =No significant difference from 70")
```

^**INSIGHTS**
  * MEAN READING SCORE IS SLIGHTLY LOWER THEN 70

  * **P-VALUE** **(0.071)>0.05** DIFFERENCE IS NOT SIGNIFICANT.

  * NO STRONH EVIDENCE THAT **READING SCORES** DIFFER FROM **70**

2. **Check if mean Writing score = 70**

```python
writing_scores = student['Writing Score']
z_stat, p_value = ztest(writing_scores, value=70)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H₀  Mean math score is significantly different from 70")
else:
    print("Fail to reject H₀  No significant difference from 70")
```

^ **INSIGHTS**
   * MEAN **WRITING SCORE** IS LOWER THAN **70**

   * P-VALUE IS VERY SMALL (<0.05) -DIFFERENCE IS STATISTICALLY SIGNIFICANT

   * WRITING SCORES ARE **SIGNIFICANTLY DIFFERENT** FROM **70**

^^**TWO-SAMPEL Z-TEST** 

1. **Compare male vs female math scores**

```python
male_scores = student[student['Gender'] == 'male']['Math Score']
female_scores = student[student['Gender'] == 'female']['Math Score']
z_stat, p_value = ztest(male_scores, female_scores)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H₀ = Significant difference between male and female math scores")
else:
    print("Fail to reject H₀ = No significant difference")
```

^ **INSIGHTS**
   * MALE AND FEMALE STUDENTS HAVE NOTICEABLY DIFFERENT AVERAGE MATH SCORES

   * THE HIGH Z-STATISTIC **(5.38)** SHOWS THE DIFFERENCE IS STRONG AND UNLIKELY DUE TO CHANCE

   * P-VALUES **(7.31e-08)** IS FAR BELOW 0.05,MEANING THE DIFFERENCE

   * THE NULL HYPOTHESIS (NO DIFFERENCE) IS REJECTED WITH HIGH CONFIDENCE

2. **Compare male vs female Reading scores**

```python
male_scores = student[student['Gender'] == 'male']['Reading Score']
female_scores = student[student['Gender'] == 'female']['Reading Score']
z_stat, p_value = ztest(male_scores, female_scores)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H₀ = Significant difference between male and female math scores")
else:
    print("Fail to reject H₀ = No significant difference")
```

^ **INSIGHTS**
   * THE **NEGATIVE Z-STATISTIC** MEANS FEMALES SCORED HIGHER ON AVERAGE THEN MALES IN READING

   * **GENDER** HAS A MEASURABLE EFFECT ON **READING SCORES**

   * THE DIFFERENCE IS **STATISTICALLY SIGNIFICANT** AND **UNLIKELY TO BE RANDOM VARIATION** 

3. **Compare male vs female writing scores**

```python
male_scores = student[student['Gender'] == 'male']['Writing Score']
female_scores = student[student['Gender'] == 'female']['Writing Score']
z_stat, p_value = ztest(male_scores, female_scores)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H₀ = Significant difference between male and female math scores")
else:
    print("Fail to reject H₀ = No significant difference")
```

^ **INSIGHTS**
   * A **Z-TEST** WAS USED TO COMPARE **WRITING SCORES** BETWEEN **MALE AND FEMALE STUDENTS**

   * **Z-STATISTIC** -9.98 (NEGATIVE VALUES MEANS FEMALES SCORED HIGHER ON AVERAGE THEN MALES)
